```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')
users.head(10)
```

```python
df = pd.DataFrame(users)
df1 = df.groupby(['occupation']).age.mean()
print(df1)
```

```python
import matplotlib.pyplot as plt
```

```python
US_marriage_divorce = pd.read_csv('us-marriages-divorces-1867-2014.csv')
US_marriage_divorce.info()
```

```python
years = US_marriage_divorce['Year']
marriages = US_marriage_divorce['Marriages_per_1000']
divorces = US_marriage_divorce['Divorces_per_1000']
US_marriage_divorce = plt.figure(figsize= (16,8))
US_marriage_divorce = plt.plot(years, marriages, label ='Number of Marriages')
US_marriage_divorce = plt.plot(years, divorces, label = 'Number of Divorces')
US_marriage_divorce = plt.title("Number of marriages and divorces per capita in the U.S. between 1867 and 2014", fontsize=16)
US_marriage_divorce = plt.xlabel("Years", fontsize=14)
US_marriage_divorce = plt.legend(fontsize = 12, loc = "upper left")
US_marriage_divorce = plt.grid(True)
US_marriage_divorce
```

```python
Graph2 = pd.read_csv('us-marriages-divorces-1867-2014.csv')
Graph2 = Graph2.loc[Graph2['Year'].isin([1900,1950,2000]), ['Year', 'Marriages_per_1000', 'Divorces_per_1000']]

Graph2 = Graph2[Graph2['Year'].apply(lambda x: x in [1900, 1950, 2000])]
years = Graph2['Year']
marriages = Graph2['Marriages_per_1000']
divorces = Graph2['Divorces_per_1000']
Graph2 = plt.figure(figsize= (16,8))
Graph2 = plt.bar(years, marriages, label ='Number of Marriages')
Graph2 = plt.bar(years, divorces, label = 'Number of Divorces')
Graph2 = plt.title("Number of marriages and divorces per capita in the U.S. between 1990,1950 and 2000", fontsize=16)
Graph2 = plt.xlabel("Years", fontsize=14)
Graph2 = plt.legend(fontsize = 12, loc = "upper left")
Graph2
```

```python
ScatterPlot = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
ScatterPlot
```

```python
years = ScatterPlot['Year']
arcade = ScatterPlot['Total Arcade Revenue (billions)']
computerAward = ScatterPlot['Computer Science Doctorates Awarded (US)']
ScatterPlot = plt.figure(figsize= (16,8))
ScatterPlot = plt.scatter(years, arcade, label ='Total Arcade Revenue (billions)')
ScatterPlot = plt.scatter(years, computerAward, label = 'Computer Science Doctorates Awarded (US)')
ScatterPlot = plt.title("Total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.", fontsize=16)
ScatterPlot = plt.xlabel("Years", fontsize=14)
ScatterPlot = plt.legend(fontsize = 12, loc = "upper left")
ScatterPlot = plt.grid(True)
ScatterPlot
```
